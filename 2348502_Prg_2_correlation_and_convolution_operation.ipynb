```python
# importing the modules needed
import cv2
import numpy as np
from scipy.signal import convolve
from google.colab.patches import cv2_imshow
```

```python
# Reading the image
image = cv2.imread('clowns.png')
```

```python
# Define the identity kernel
identity_kernel = np.array([[1, 0, 0],
                            [0, 1, 0],
                            [0, 0, 1]])

# Apply the identity kernel to the image using filter2D
identity_output = cv2.filter2D(image, -1, identity_kernel)

# Display the original and output images
cv2_imshow(image)
cv2_imshow(identity_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1, 8, -1],
                                  [-1, -1, -1]])
edge_detection_output = cv2.filter2D(image, -1, edge_detection_kernel)
cv2_imshow(image)
cv2_imshow (edge_detection_output)
```

```python
sharpen_kernel = np.array([[0, -1, 0],
                            [-1, 5, -1],
                            [0, -1, 0]])
sharpen_output = cv2.filter2D(image, -1, sharpen_kernel)

cv2_imshow(image)
cv2_imshow( sharpen_output)
```

```python
box_blur_kernel = np.ones((3, 3)) / 9
box_blur_output = cv2.filter2D(image, -1, box_blur_kernel)

cv2_imshow(image)
cv2_imshow(box_blur_output)
```

```python
gaussian_blur_kernel = np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]]) / 16

gaussian_blur_output = cv2.filter2D(image, -1, gaussian_blur_kernel)
cv2_imshow(image)
cv2_imshow(gaussian_blur_output)
```
