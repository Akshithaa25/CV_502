```python
# importing the modules needed
import cv2
import numpy as np
from scipy.signal import convolve
from google.colab.patches import cv2_imshow
```

```python
# Reading the image
image = cv2.imread('clowns.png')
cv2_imshow(image)
```

```python
# Define the identity kernel
identity_kernel = np.array([[1, 0, 0],
                            [0, 1, 0],
                            [0, 0, 1]])

# Apply the identity kernel to the image using filter2D
identity_output = cv2.filter2D(image, -1, identity_kernel)

# Display the original and output images
cv2_imshow(image)
cv2_imshow(identity_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1, 8, -1],
                                  [-1, -1, -1]])
edge_detection_output = cv2.filter2D(image, -1, edge_detection_kernel)

cv2_imshow (edge_detection_output)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
|# Define the kernels


edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1, 8, -1],
                                  [-1, -1, -1]])

sharpen_kernel = np.array([[0, -1, 0],
                            [-1, 5, -1],
                            [0, -1, 0]])

box_blur_kernel = np.ones((3, 3)) / 9

gaussian_blur_kernel = np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]]) / 16

# Apply the kernels to the image using filter2D
identity_output = cv2.filter2D(image, -1, identity_kernel)
edge_detection_output = cv2.filter2D(image, -1, edge_detection_kernel)
sharpen_output = cv2.filter2D(image, -1, sharpen_kernel)
box_blur_output = cv2.filter2D(image, -1, box_blur_kernel)
gaussian_blur_output = cv2.filter2D(image, -1, gaussian_blur_kernel)

# Apply the kernels to the image using convolve
identity_output_convolve = convolve(image, identity_kernel, mode='same')
edge_detection_output_convolve = convolve(image, edge_detection_kernel, mode='same')
sharpen_output_convolve = convolve(image, sharpen_kernel, mode='same')
box_blur_output_convolve = convolve(image, box_blur_kernel, mode='same')
gaussian_blur_output_convolve = convolve(image, gaussian_blur_kernel, mode='same')

# Display the original and output images
cv2_imshow('Original Image', image)
cv2_imshow('Identity Output', identity_output)
cv2_imshow('Edge Detection Output', edge_detection_output)
cv2_imshow('Sharpen Output', sharpen_output)
cv2_imshow('Box Blur Output', box_blur_output)
cv2_imshow('Gaussian Blur Output', gaussian_blur_output)
cv2_imshow('Identity Output Convolve', identity_output_convolve)
cv2_imshow('Edge Detection Output Convolve', edge_detection_output_convolve)
cv2_imshow('Sharpen Output Convolve', sharpen_output_convolve)
cv2_imshow('Box Blur Output Convolve', box_blur_output_convolve)
cv2_imshow('Gaussian Blur Output Convolve', gaussian_blur_output_convolve)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
