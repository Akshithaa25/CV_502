```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Load the input image
image_path = ("/content/clowns.png")
input_image = cv2.imread(image_path)
```

```python
# Box filter (Mean filter)
# mean_blur = cv2.blur(gray_image, (5, 5))


def user_defined_mean_filter(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    return cv2.filter2D(image, -1, kernel)

# User-defined Box filter (Mean filter)
mean_blur = user_defined_mean_filter(input_image, 5)

plt.figure(figsize=(15, 10))
plt.subplot(331), plt.imshow(input_image, cmap='gray'), plt.title('Original Image')
plt.subplot(332), plt.imshow(mean_blur, cmap='gray'), plt.title('Mean Blur')
```

```python
# Gaussian Blur
# gaussian_blur= cv2.GaussianBlur(gray_image, (5, 5), 0)


def user_defined_gaussian_blur(image, kernel_size):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

gaussian_blur= user_defined_gaussian_blur(input_image, 5)

plt.figure(figsize=(15, 10))
plt.subplot(334), plt.imshow(input_image, cmap='gray'), plt.title('Original Image')
plt.subplot(335), plt.imshow(gaussian_blur, cmap='gray'), plt.title(' Gaussian Blur')
```
