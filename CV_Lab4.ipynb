```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the input image
image_path = ("/content/img 1.png")
input_image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Built-in functions:

# Box filter (Mean filter)
mean_blur_builtin = cv2.blur(gray_image, (5, 5))

# Gaussian Blur
gaussian_blur_builtin = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Laplace filter (Sharpening)
laplace_builtin = cv2.Laplacian(gray_image, cv2.CV_64F)

# User-defined functions:

def user_defined_mean_filter(image, kernel_size):
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    return cv2.filter2D(image, -1, kernel)

def user_defined_gaussian_blur(image, kernel_size):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

def user_defined_laplace_filter(image):
    laplace_kernel = np.array([[0, 1, 0],
                               [1, -4, 1],
                               [0, 1, 0]], dtype=np.float32)
    return cv2.filter2D(image, -1, laplace_kernel)

# User-defined Box filter (Mean filter)
mean_blur_user_defined = user_defined_mean_filter(gray_image, 5)

# User-defined Gaussian Blur
gaussian_blur_user_defined = user_defined_gaussian_blur(gray_image, 5)

# User-defined Laplace filter
laplace_user_defined = user_defined_laplace_filter(gray_image)

# Display the results:

plt.figure(figsize=(15, 10))

plt.subplot(331), plt.imshow(gray_image, cmap='gray'), plt.title('Original Image')
plt.subplot(332), plt.imshow(mean_blur_builtin, cmap='gray'), plt.title('Built-in Mean Blur')
plt.subplot(333), plt.imshow(mean_blur_user_defined, cmap='gray'), plt.title('User-defined Mean Blur')

plt.subplot(334), plt.imshow(gray_image, cmap='gray'), plt.title('Original Image')
plt.subplot(335), plt.imshow(gaussian_blur_builtin, cmap='gray'), plt.title('Built-in Gaussian Blur')
plt.subplot(336), plt.imshow(gaussian_blur_user_defined, cmap='gray'), plt.title('User-defined Gaussian Blur')

plt.subplot(337), plt.imshow(gray_image, cmap='gray'), plt.title('Original Image')
plt.subplot(338), plt.imshow(laplace_builtin, cmap='gray'), plt.title('Built-in Laplace Filter')
plt.subplot(339), plt.imshow(laplace_user_defined, cmap='gray'), plt.title('User-defined Laplace Filter')

plt.show()
```
