####Python program to read image using PIL module


```python
#Using PIL : PIL is the Python Imaging Library
# importing PIL
#Python Imaging Library
from PIL import Image

# Read image
img = Image.open('leaf.jpg')
```

```python
# Output Images
display(img)
#We use display to show image in PIL module
```

```python
# prints format of image
print(img.format)
# prints Format of the image
# Eg- RGB, RFBA, CMYK, BGRetc.
print(img.mode)
print("Filename: ", img.filename)
print("Size: ", img.size)
# Prints the width of the image (in pixels)
print("Width: ", img.width)
# Prints the height of the image (in pixels)
print("Height: ", img.height)
```

#Converting the given images in to different formats (For ex: *.jpeg, *.jpg, *.jpe, , *.png, tiff, *.tif, e.t.c)

```python
# Loading the image
image = Image.open('code.jpg')

# Specifying the RGB mode to the image
image = image.convert('RGB')

# Converting an image from PNG to JPG format
image.save("converted-png-image.png")
print("Image successfully converted!")
```
